
# Scatterplot


```python
#syntax is same, instead of hist, we use scatter

plt.figure(figsize=(8,8))
plt.scatter(data1['horsepower'],data1['mpg'],s=10) #s=size of marker
plt.title('Plot of HP vs MPG',fontsize = 16) 
plt.xlabel('HP')  #adding x and y labels
plt.ylabel('MPG')
```

```python
#From the graph we understand that there is negative corelation. as HP increases, MPG is decreasing
#to change the colour, colour in histogram, but c in scatter, and make marker from dot to triangle 
```

```python
plt.figure(figsize=(8,8))
plt.scatter(data1['horsepower'],data1['mpg'],s=10,c='Red',marker ='^') #s=size of marker
plt.title('Plot of HP vs MPG',fontsize = 16) 
plt.xlabel('HP')  #adding x and y labels
plt.ylabel('MPG')
```

```python
#scatterplot dispalcement vs mpg
plt.figure(figsize=(8,8))
plt.scatter(data1['displacement'],data1['mpg'],s=10,c='Red',marker ='^') #s=size of marker
plt.title('Plot of Disp vs MPG',fontsize = 16) 
plt.xlabel('Displacement')  #adding x and y labels
plt.ylabel('MPG')
```

```python
# to display two of these graphs in a single output up and down use subplot


plt.figure(figsize=(8,8))

plt.subplot(2,1,1)  #In a figure area, it will divide it into a 2*1 matrix and plot first one 

plt.scatter(data1['horsepower'],data1['mpg'],s=10,c='Red',marker ='^') 
plt.title('Plot of HP vs MPG',fontsize = 16) 
plt.xlabel('HP')  #adding x and y labels
plt.ylabel('MPG')

plt.subplot(2,1,2)  #In a figure area, it will divide it into a 2*1 matrix and plot in second one

plt.scatter(data1['displacement'],data1['mpg'],s=10,c='blue',marker ='^') 
plt.title('Plot of Disp vs MPG',fontsize = 16) 
plt.xlabel('Displacement')  #adding x and y labels
plt.ylabel('MPG')

plt.tight_layout()  #to give sufficient padding to avoid overlap of two plots
```

# Strip Plots

Strip plot is a scatterplot where one of the variables is catergorical. 
They can be combined with other plots to get more meaningful data 

A strip plot is a graphical data anlysis technique for summarizing a univariate data set. 
The strip plot consists of:
Horizontal axis = the value of the response variable;
Verticalal axis = all values are set to 1.





```python
sns.stripplot(x='mpg',data=data1)
```

# Swarm plots

Swarm Plots, also called beeswarm plots, are similar to strip plots in that they plot all of the data points. Unlike strip plots, swarm plots attempt to avoid obscuring points by calculating non-overlapping positions instead of adding random jitter.
It is a good complement to box plot or a violin plot incase you wantto show all observations. 

```python
sns.swarmplot(x='mpg',data=data1)
```

```python
sns.swarmplot(x= 'cylinders',y='mpg',data=data1)
```

```python
sns.stripplot(x= 'cylinders',y='mpg',data=data1)
```

# Violin plot 

A violin plot is a hybrid of a box plot and a kernel density plot, which shows peaks in the data. It is used to visualize the distribution of numerical data. Unlike a box plot that can only show summary statistics, violin plots depict summary statistics and the density of each variable.

```python
sns.violinplot(x= 'cylinders',y='mpg',data=data1)
```

# countplot

A count plot can be thought of as a histogram across a categorical, instead of quantitative, variable. 

The basic API and options are identical to those for barplot() , so you can compare counts across nested variables.


```python
sns.countplot(x='origin',data = data1)
```
